##Generate Fish

```
import random

#given that you wish to separate the list from the function
fishes = list()

def add_multiple(fish, number):
  #a fish should be an object
  fishes.extend([fish for i in range(number)])

add_multiple('flopper', 10)
add_multiple('midas', 1)
#and many others

[print(random.choice(fishes)) for x in range(10)]
#None
```

##Generate Size_normal

```
def simulate_normal(sd, mean, alfa, size):
  #simulated probability distribution of one stock
  import pandas as pd
  from scipy.stats import skewnorm
  import matplotlib.pyplot as plt

  #invertire il segno di alfa
  x = skewnorm.rvs(-alfa, size=size) 
  def calc(k, sd, mean):
    return (k*sd)+mean
  x = calc(x, sd, mean) #standard distribution

  fig = plt.figure(figsize = (14,7))
  ax = fig.gca()
  pd.DataFrame(x).hist(ax=ax, bins=100)
  return pd.DataFrame(x)

  #pick one random number from the distribution
  #formally I would use cdf, but I just have to pick randomly from the 1000000 samples
x = simulate_normal(14, 40, -4, 1000000)
x
```

##Create List of Fishes

```
fish_list = list()
class Fish():
  def __init__(self):
    self.species = random.choice(fishes)
    self.size = random.choice(x[0])
    fish_list.append(self)

  def print_data(self):
    print(self.species, self.size)

  def cheat(self, species, size):
    self.species = species
    self.size = size

[Fish() for _ in range(10)]
fish_list
```

```
for _ in fish_list:
  _.print_data()
```
